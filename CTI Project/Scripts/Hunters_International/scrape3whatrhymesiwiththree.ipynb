```python


# Create a reverse mapping from data descriptions to data types
description_to_type = {}
for data_type, descriptions in mapped_data.items():
    for description in descriptions:
        description_to_type[description.title()] = data_type

# Read the CSV file and process the rows
with open('hunters_international_filtered.csv', 'r') as f:
    reader = csv.reader(f)
    header = next(reader)

    # Extract the data descriptions from the header
    data_descriptions = header[6:]

    # Create a new header with the data types
    new_data_types = list(mapped_data.keys())
    new_header = header[:6] + new_data_types

    # Process the rows to map data descriptions to data types
    new_rows = []
    for row in reader:
        new_row = row[:6]
        data_description_flags = row[6:]

        # Initialize the new data type columns with '0'
        data_type_flags = ['0'] * len(new_data_types)

        # Map the data descriptions to data types
        for i, flag in enumerate(data_description_flags):
            if flag == '1':
                description = data_descriptions[i]
                if description in description_to_type:
                    data_type = description_to_type[description]
                    data_type_index = new_data_types.index(data_type)
                    data_type_flags[data_type_index] = str(
                        int(data_type_flags[data_type_index]) + 1)

        new_row.extend(data_type_flags)
        new_rows.append(new_row)

# Write the updated rows to a new CSV file
with open('hunters_international_filtered_categorized.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(new_header)
    writer.writerows(new_rows)
```
