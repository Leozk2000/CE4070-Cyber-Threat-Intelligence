```python
import zipfile
import re
from bs4 import BeautifulSoup


company_name_pattern = r'<h1 class=title>(.*?)</h1>'
revenue_pattern = r'<em>Revenue</em>: (.*?)<'
data_volume_pattern = r'<em>Data Volume</em>: (.*?)<'
date_pattern = r'article:published_time" content="([\d\-T:+]+)"'
data_description_pattern = r'<ul>(.*?)</ul>'  
countryIndustry_pattern = r'<li><a href=/tags/.*?>(.*?)</a></li>'


extracted_data = []


def extract_data(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')


    company_name = re.search(company_name_pattern, html_content)
    revenue = re.search(revenue_pattern, html_content)
    data_volume = re.search(data_volume_pattern, html_content)
    attack_date = re.search(date_pattern, html_content)
    data_description = re.search(data_description_pattern, html_content, re.DOTALL)
    

    matches = re.findall(countryIndustry_pattern, html_content)
    country = matches[0] if matches else 'N/A'
    industries = matches[1:] if len(matches) > 1 else ['N/A']


    

    if data_description:
        data_description = BeautifulSoup(data_description.group(1), 'html.parser').get_text(separator=', ')


    data = {
        'company_name': company_name.group(1) if company_name else 'N/A',
        'date_of_attack': attack_date.group(1) if attack_date else 'N/A',
        'revenue': revenue.group(1) if revenue else 'N/A',
        'data_volume': data_volume.group(1) if data_volume else 'N/A',
        'data_description': data_description if data_description else 'N/A',
        'country': country,
        'industry': ', '.join(industries) if industries else 'N/A'
    }

    return data


zip_file_path = 'C:\\Users\\Sruthi\\OneDrive\\Sruthi\\Studies\\Uni\\Uni Courses-Sruthi-Zenbook\\Year 3 Sem 1\\Cyber Threat Intelligence\\Project\\bianlian scrapped txt.zip'



with zipfile.ZipFile(zip_file_path, 'r') as z:
    for file_name in z.namelist():
        with z.open(file_name) as file:
            content = file.read().decode('utf-8')
            extracted_info = extract_data(content)
            extracted_data.append(extract_data(content))
            print(f"Extracted data from file {file_name}:")
            for key, value in extracted_info.items():
                print(f"{key}: {value}")
            print("\n" + "-"*50 + "\n")

csv_columns = ['company_name', 'date_of_attack', 'revenue', 'data_volume', 'data_description', 'country', 'industry']

csv_file = 'extracted_data.csv'
try:
    with open(csv_file, 'w', newline='') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in extracted_data:
            writer.writerow(data)
except IOError:
    print("I/O error")
```
